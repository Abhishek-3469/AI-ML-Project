**PROJECT 1: IDENTIFY MULTIPLE DISEASE**

```python
#STEP 1: Import Library
import pandas as pd
```

```python
#STEP 2: Read Dataset or Dataset Load
df=pd.read_csv('https://raw.githubusercontent.com/ybifoundation/Dataset/main/MultipleDiseasePrediction.csv')
```

```python
df.head()
```

```python
df.info(verbose=True)
```

```python
df.describe()
```

```python
#STEP 3: Define y and x or Define Output and Input
df.columns
```

```python
y=df['prognosis']
x=df.drop(['prognosis'],axis=1)
```

```python
(df['prognosis'].unique())
```

```python
df['prognosis'].nunique()
```

```python
df['prognosis'].value_counts()
```

```python
#STEP 4: Train and Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2529)
```

```python
df.shape,x_train.shape,x_test.shape
```

```python
#STEP 5: Select Model
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
#STEP 6: Train Model
model.fit (x_train,y_train)
```

```python
#STEP 7: Predict
y_pred = model.predict(x_test)
```

```python
#STEP 8: Evaluate
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
